### distribution plot

```python
sns.distplot(crash_df['alcohol'],kde=False, bins=5)
```

Distplots are commonly used in exploratory data analysis to assess data quality, detect outliers, and determine the appropriateness of statistical tests and modeling techniques.
Bins define how many buckets to divide the data up into between intervals

### joint plot

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

Jointplot compares 2 distributions and plots a scatter plot by default

With kind you can create a regression line with kind='reg'

You can create a 2D KDE with kind='kde'

Kernal Density Estimation estimates the distribution of data

You can create a hexagon distribution with kind='hex'


### KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

* Use a KDE plot when you need to explore the distribution of a continuous variable, such as age, income, test scores, or temperature. It helps you understand how data is spread out.
* A KDE plot provides a visual representation of the probability density of data. It allows you to see where data points are likely to be more concentrated and where they are less concentrated.
* You can use KDE plots to compare distributions of different groups or datasets

```python
sns.pairplot(crash_df)
```

 Pair plots are used when you have several variables in your dataset and you want to examine how they relate to each other

 You can use a pair plot to quickly identify strong positive or negative correlations between variables.
 

```python
tips_df=sns.load_dataset('tips')
tips_df
sns.pairplot(tips_df,hue='sex',palette='Blues')
```

 With hue you can pass in a categorical column and the charts will be colorized.

You can use color maps from Matplotlib to define what colors to use



### Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

Plots a single column of datapoints in an array as sticks on an axis.
    
With a rug plot you'll see a more dense number of lines where the amount is most common.

This is like how a histogram is taller where values are more common

### Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=tips_df, estimator=np.var)
```

 Bar plots are commonly used when you want to compare the values or frequencies of different categories or groups

  Bar plots can be used to show the distribution of data within categories, such as histograms, where each category represents a range or bin for continuous data.

  You can use bar plots to represent data at discrete time points or intervals, making them suitable for visualizing trends over time when you have data at fixed intervals

### Count Plot

```python
sns.countplot(x='sex',data=tips_df)
```

A count plot is like a bar plot, but the estimator is counting the number of occurances

A countplot is a type of data visualization that is commonly used in data analysis and data science to display the frequency or count of categorical variables. It's a useful tool for understanding the distribution of categorical data and identifying patterns or trends within the data.

### Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

A box plot allows you to compare different variables

The box shows the quartiles of the data. The bar in the middle is the median and the box extends 1 standard deviation from the median

 The whiskers extend to all the other data aside from the points that are considered to be outliers

### Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex',split=True)
```

Violin Plot is a combination of the boxplot and KDE

While a box plot corresponds to data points, the violin plot uses the KDE estimation of the data points

Split allows you to compare how the categories compare to each other

### Stripplot

```python
plt.figure(figure=(8,5))
sns.stripplot(x='day',y='total_bill',data=tips_df,jitter=True,hue='sex',dodge=True)
```

The strip plot draws a scatter plot representing all data points where one variable is categorical. It is often used to show all observations with a box plot that represents the average distribution

Jitter spreads data points out so that they aren't stacked on top of each other

 Hue breaks data into men and women

 Dodge separates the men and women data

# swarm plot

```python
#sns.violinplot(x='day',y='total_bill',data=tips_df)
sns.swarmplot(x='day',y='total_bill',data=tips_df)
```

A swarm plot is like a strip plot, but points are adjusted so they don't overlap

 It looks like a combination of the violin and strip plots
