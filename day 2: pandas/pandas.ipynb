## Single Filtering


### First: Using Simple Condition


```python
ranking = df[df['ranking']<5]
ranking
```

### Second: Using Anonymous Function[map and lambda]


```python
above_5 = df[df['ranking'].map(lambda x:x<5)]
above_5
```

## Muliple Filtering


### 1.Using Simple Logical Conditions


```python
above_100_mystery = df[(df['ranking']<5) & (df['Genre']=='Comedy')]
above_100_mystery
```

### 2. Using Map and Lambda


```python
less_than_5_Action = df[df['ranking'].map(lambda x:x<5) & df['Genre'].map(lambda movie:movie=="Action")]
less_than_5_Action
```

## Dropna

```python
df.dropna(axis=1).head(2)
```

"dropna" is a function that helps you remove rows or columns from your DataFrame where there are missing values

To remove rows with missing values, you can do something like this:


```python
df.dropna(axis=0, inplace=True)
```

The "axis=0" part tells Pandas to look at rows, and "inplace=True" means that the changes will be made directly to your DataFrame

To remove columns with missing values, you can do something like this:

```python
df.dropna(axis=1, inplace=True)
```

 "axis=1" tells Pandas to look at columns. Again, "inplace=True" makes the changes directly in your DataFrame.

```python
df.dropna(axis=1).columns
```

### Fillna

"fillna" works by letting you specify what you want to put in place of missing values. You can fill in missing values with a specific number, a word, the mean or median of a column, or any other value that makes sense in the context of your data.

To fill missing values with a specific number (let's say, 0):

```python
df.fillna(0, inplace=True)
```

This code replaces all the "NaN" values in your DataFrame df with the number 0. The inplace=True part means you're making the change directly in your DataFrame.

### Groupby

```python
df = pd.read_csv('modified.csv')

df['count'] = 1

df.groupby(['Type 1', 'Type 2']).count()['count']
```

```python
df = pd.DataFrame({'Anime/Manga': ['One Piece', 'One Piece',
                              'AOT', 'AOT','Gintama','Gintama'],
                   'Episodes/Chapter': [1015,1047,75,129,366,704]})
```

```python
df.groupby(['Anime/Manga']).mean()
```

 This function allows us to group and analyze data in a more structured way, making complex operations on data easier.

 The "groupby" function helps us group data in a DataFrame based on a specific column or multiple columns. It's important because it enables us to perform various operations, such as aggregations, on those groups separately

```python
df.groupby(['Anime/Manga']).max()
```

```python
df.groupby(['Anime/Manga']).agg(['mean','max','min','count'])
```
