# MATPLOTLIB

Matplotlib - An object-oriented plotting python library. Matplotlib is the Data Visualization tool in Python, in which you can plot amazing charts. The major feature of Matplotlib is ability to plot graphics backends, which supports dozen of charts that includes Bar chart, Pie Chart, Histogram, Scatter Plot, Line Charts and many more.

### Loading Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

### Creating figure and axes

```python
fig = plt.figure()
ax = plt.axes()
plt.show() 
```

Figure: A Figure in Matplotlib is like a blank canvas or a window that contains all the elements of a plot or a chart.

Axes: Axes in Matplotlib are the actual plotting area or the individual plots that you create inside a Figure

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y,label='scale=1',color='red',linewidth=2,marker='.',markersize=10)
plt.title('first graph',fontdict={"fontsize":10})
plt.xlabel('x')
plt.ylabel('y')
plt.xticks([1,2,3,4])
plt.yticks([2,4,6,8])
plt.legend()
plt.show()
```

```python
x=[0,1,2,3]
y=[0,2,4,6]
plt.plot(x,y,'b^--',label='2x')
x2=np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,'r.--',label='x2')
plt.title('first graph',fontdict={"fontsize":10})
plt.xlabel('x')
plt.ylabel('y')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```

```python
x=[0,1,2,3]
y=[0,2,4,6]
plt.figure(figsize=(6,3),dpi=300)
plt.plot(x,y,'b^--',label='2x')
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:5],x2[:5]**2,'r',label='x2')
plt.plot(x2[4:],x2[4:]**2,'r.--')
plt.title('first graph',fontdict={"fontsize":10})
plt.xlabel('x')
plt.ylabel('y')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```

### Bar Plot


```python
labels=['A','B','C']
Values=[1,4,2]
plt.bar(Values,labels)
```

Bar graphs are the pictorial representation of data (generally grouped), in the form of vertical or horizontal rectangular bars, where the length of bars are proportional to the measure of data.

```python
labels=['A','B','C']
Values=[1,4,2]
bars=plt.bar(labels,Values)
patterns=['/','O','*']
for bars in bars:
    bars.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))
plt.show
```

```python
labels=['A','B','C']
Values=[1,4,2]
bars=plt.bar(labels,Values)
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show
```

### Histogram Plot


```python
heights = np.array([130,146,155,157,159,171,176,168,188,180,165,166,164,168,174,171,176,175,166,177,180,189,190,162,154,159,150])
plt.hist(heights,bins=10)
```

Histogram is usually used to plot numeric values.



```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

### Pie Plot

```python
marks = np.array([92,91,90,85,81])
subjects = ["Computer","Maths","Electronics","English","Management"]
plt.pie(marks,labels=subjects,shadow=True,explode=[0,0,0.2,0,0.15],autopct="%0.3f%%")
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

### Scatter Plot


```python
x = [12,45,23,10,30]
y = [20,17,34,3,10]
plt.scatter(x,y,color="red")
plt.show()
```

Scatter Plot is used to compare between two features, i.e., to check relationship between two or more variable.



### Subplot

```python
anime_names = ["One Piece","Naruto","Bleach","Gintama","Fairy Tail"]
episodes = [1000,720,366,360,180]

plt.subplot(2,1,1)
plt.bar(anime_names,episodes)
plt.ylabel("Total Episodes")

plt.subplot(2,1,2)
plt.bar(anime_names,episodes,color=["red","yellow","black","green","blue"])
plt.xlabel("Anime Names")
plt.ylabel("Total Episodes")
plt.show()
```

Subplot is used to plot multiple graphs in one single figure windows by sub-dividing into sub categories.



### Box Plot

```python

plt.figure(figsize=(5,8), dpi=100)
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')  
plt.show()
```
